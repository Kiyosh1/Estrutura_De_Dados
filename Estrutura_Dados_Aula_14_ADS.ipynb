Implementação do Selection Sort para ordenação de elementos em uma coluna com pandas DataFrame

```python
def selection_sort_column(df, column_name):
    """
    Ordena uma coluna específica de um DataFrame usando o algoritmo Selection Sort.

    Parâmetros:
    df : pd.DataFrame - O DataFrame contendo a coluna a ser ordenada.
    column_name : str - O nome da coluna a ser ordenada.

    Retorna:
    pd.DataFrame - O DataFrame com a coluna ordenada.
    """
    # Criar uma cópia do DataFrame para não modificar o original
    df_sorted = df.copy()
    n = len(df_sorted[column_name])

    # Aplicação do Selection Sort na coluna especificada
    for i in range(n - 1):
        min_idx = i
        for j in range(i + 1, n):
            if df_sorted.loc[j, column_name] < df_sorted.loc[min_idx, column_name]:
                min_idx = j
        # Troca os valores na posição i e min_idx
        if min_idx != i:
            temp = df_sorted.loc[i, column_name]
            df_sorted.loc[i, column_name] = df_sorted.loc[min_idx, column_name]
            df_sorted.loc[min_idx, column_name] = temp

    return df_sorted
```

```python
# Exemplo de uso:
data = {'A': [3, 1, 4, 1, 5], 'B': [9, 2, 6, 5, 3]}
dfx = pd.DataFrame(data)
print("DataFrame Original:")
print(dfx)
```

```python
# Ordena a coluna 'A'
df_sorted = selection_sort_column(dfx, 'B')
print("\nDataFrame com a coluna 'A' ordenada:")
print(df_sorted)
```

Implementação do bubble_sort para ordenação de elementos em uma coluna com pandas DataFrame

```python
def bubble_sort_column(df, column_name):
    """
    Ordena uma coluna específica de um DataFrame usando o algoritmo Bubble Sort.

    Parâmetros:
    df : pd.DataFrame - O DataFrame contendo a coluna a ser ordenada.
    column_name : str - O nome da coluna a ser ordenada.

    Retorna:
    pd.DataFrame - O DataFrame com a coluna ordenada.
    """
    # Criar uma cópia do DataFrame para não modificar o original
    df_sorted = df.copy()
    n = len(df_sorted[column_name])

    # Aplicação do Bubble Sort na coluna especificada
    for i in range(n - 1):
        for j in range(n - 1 - i):
            if df_sorted.loc[j, column_name] > df_sorted.loc[j + 1, column_name]:
                # Troca os valores nas posições j e j+1
                temp = df_sorted.loc[j, column_name]
                df_sorted.loc[j, column_name] = df_sorted.loc[j + 1, column_name]
                df_sorted.loc[j + 1, column_name] = temp

    return df_sorted
```

```python
# Exemplo de uso:
data = {'A': [3, 9, 4, 1, 5], 'B': [9, 2, 6, 5, 3]}
dfx = pd.DataFrame(data)
print("DataFrame Original:")
print(dfx)
```

```python
# Ordena a coluna 'A'
df_sorted = bubble_sort_column(dfx, 'A')
print("\nDataFrame com a coluna 'A' ordenada:")
print(df_sorted)
```
