#Exercício - Aula 7

## Questão 1

Implemente uma TAD Lista Posicional em Python que permite inserir elementos na primeira e última posição, e também permite navegar pela lista.

(1) Crie uma lista posicional com os seguintes elementos, nesta ordem:

1. Adicione o número 10 no início.
2. Adicione o número 20 no final.
3. Adicione o número 30 no final.
4. Adicione o número 25 antes do número 30.
5. Adicione o número 15 após o número 10.

(2) Criar um método apresente os elementos dentro da lista posicional.
